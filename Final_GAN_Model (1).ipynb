#**Data Pre-processing**

```
from music21 import converter, instrument, note, chord
```

```
def get_files(dataset_path):
  file_list = []
  for file in glob.glob(dataset_path):
    midi = converter.parse(file)
    file_list.append(midi)
  print("Parsing is done...")
  return file_list

def extract_allnotes(file_list):
    notes = []
    notes_on_eachsong = None
    for each_song in file_list:
        if instrument.partitionByInstrument(each_song) is not None:
            groups = instrument.partitionByInstrument(each_song)
            notes_on_eachsong = groups.parts.recurse() 
          
        
        else: # file has notes in a flat structure
            print("yes")
            notes_on_eachsong = each_song.flat.notes

        for element in notes_on_eachsong:
            if isinstance(element, note.Note):
                notes.append(str(element.pitch))
            elif isinstance(element, chord.Chord):
                notes.append('.'.join(str(n) for n in element.normalOrder))

    return notes
```

```
file_list = get_files("Classic Piano Music/*.mid")
notes = extract_allnotes(file_list)
```

```
def note_id(notes):
  unique_notes = list(set(notes))
  unique_notes.sort()
  ids = {unique_notes[i]:i for i in range(len(unique_notes))}
  print(ids)
  return ids, unique_notes
```

```
def input_sequence(notes):
  n = 100   #change seq length here
  sequences_100 = []
  i = 0
  while(i<len(notes)):
    seq = notes[i:i+n].copy()
    sequences_100.append(seq)
    i+=100  #change seq length here too

  #drop the last sequence, if it has less than 100 notes.
  if len(sequences_100[-1])<100:
    sequences_100 = sequences_100[0:len(sequences_100)-1]
  return sequences_100
```

```
def note_normalizer(sequence, unique_notes):
  for i in range(len(sequence)):
    for j in range(len(sequence[i])):
      sequence[i][j] = sequence[i][j]/float(len(unique_notes))
  return sequence
```

```
def make_midi(new_song):
  song = []
  i = 0
  time_from_start = 0
  while (i<len(new_song)):
    parts_of_element = new_song[i].split('.')
    print(parts_of_element)
    if len(parts_of_element)==1 and parts_of_element[0][0].isalpha():
      add_note = note.Note(new_song[i])
      add_note.offset = time_from_start
      add_note.storedInstrument = instrument.Piano()
      song.append(add_note)
    else:
      add_part = note.Note()
      group = []
      for part in parts_of_element:
        add_part.storedInstrument = instrument.Piano()
        #add_part = note.Note(int(part))

        group.append(add_part)
      add_chord = chord.Chord(group)
      add_chord.offset = time_from_start
      song.append(add_chord)
    time_from_start += 0.5
    i+=1
  
  midi_song = stream.Stream(song)
  print(midi_song.elements)
  midi_song.write('midi', 'GAN new song.mid')
        


    



    
```

#Proposed Model: Bi-LSTM GAN with Sequential Self-Attention

```
from keras.optimizers import Adam
from keras.layers import Input
from keras.models import Model
from keras.layers import Reshape
from keras.layers import Flatten
```

```
from keras.models import Sequential
from keras.layers import LSTM, Bidirectional, Dense, Dropout, Activation, BatchNormalization
from music21 import stream
```

```
from keras_self_attention import SeqSelfAttention
from keras.layers import GaussianNoise
```

```
latent_dim = 500
```

```
# MAIN 
def Generator():
    model = Sequential()
    model.add(Dense(256, input_dim=latent_dim, activation = "relu"))
    #model.add(BatchNormalization(momentum=0.4))
    model.add(Dense(512, activation = "relu"))
    #model.add(BatchNormalization(momentum=0.4))
    model.add(Dense(800, activation = "relu"))
    #model.add(BatchNormalization(momentum=0.4))
    model.add(Dense(256, activation = "relu"))
    model.add(BatchNormalization(momentum=0.4))
    model.add(Dense(np.prod((100,1)), activation='tanh'))
    model.add(Reshape((100,1)))
    model.summary()
    return model
```

```
generator = Generator()
generator.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```
import keras
```

```
def Discriminator():
    model = Sequential()

    model.add(LSTM(512, input_shape=(100,1), return_sequences=True))
    model.add(Bidirectional(LSTM(512, return_sequences=True)))
    model.add(GaussianNoise(0.1))
    model.add(SeqSelfAttention(attention_activation="sigmoid"))
    model.add(Dropout(0.2))
    
    model.add(Bidirectional(LSTM(512,return_sequences=True)))
    model.add(SeqSelfAttention(attention_activation = "sigmoid"))
    model.add(Dropout(0.2))
    
    model.add(Bidirectional(LSTM(512,return_sequences=True)))
    model.add(SeqSelfAttention(attention_activation="sigmoid"))
    model.add(Dense(512, activation="relu"))
    model.add(Dropout(0.2))
    
  
    model.add(Dense(250, activation = "relu"))
    model.add(Dense(64, activation = "relu"))
    model.add(Dense(1, activation='sigmoid'))

    model.summary()
    return model
```

```
discriminator = Discriminator()
discriminator.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```
 def plot_loss(discriminator_losses, generator_losses):
        plt.plot(discriminator_losses, c='green')
        plt.plot(generator_losses, c='orange')
        
        plt.xlabel('Epoch Number')
        plt.ylabel('Loss Value')

        plt.legend(['Discriminator', 'Generator'])
        plt.title("GAN: Loss Vs Epoch")

        plt.savefig('GAN_Loss_per_Epoch_final.png', transparent=True)
        plt.close()
```

```
def plot_accuracy(discriminator_accuracy):
        plt.plot(discriminator_accuracy, c='green')
        
        plt.xlabel('Epoch Number')
        plt.ylabel('Accuracy')

        plt.legend(['Discriminator'])
        plt.title("GAN: Accuracy Vs Epoch")

        plt.savefig('GAN_Accuracy_per_Epoch_final.png', transparent=True)
        plt.close()
```

```
stacked_gan = Sequential()
stacked_gan.add(generator)
discriminator.trainable = False
stacked_gan.add(discriminator)
```

```
stacked_gan.compile(loss = "binary_crossentropy", optimizer = "adam", metrics = ["accuracy"])
```

```
batch_size = 32
num_of_batches = int(in_array.shape[0]/batch_size)
```

```
num_of_batches
```

```
real = np.full((batch_size, 1), 1)   
fake = np.full((batch_size, 1), 0)
D_losses = []
G_losses = []
D_acc = []
G_acc = []
```

```
for epoch in range(350):

            # Training the discriminator
            # Select a random batch of note sequences
            i = 0
            while (i<2):

              idx = np.random.randint(0, in_array.shape[0], batch_size)
              real_seqs = in_array[idx]

              
              noise = np.random.normal(0, 1, (batch_size, latent_dim))
              
              # Generate a batch of new note sequences
              gen_seqs = generator.predict(noise)

              # Train the discriminator
              d_loss_real = discriminator.train_on_batch(real_seqs, real)
              d_loss_fake = discriminator.train_on_batch(gen_seqs, fake)
              d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)

              if i == 0:
                #  Training the Generator
                noise = np.random.normal(0, 1, (batch_size, latent_dim))

                # Train the generator 
                g_loss = stacked_gan.train_on_batch(noise, real)
              i+=1

            # Print the progress and save into loss lists
            if epoch % 1 == 0:
              #print ("%d [D loss: %f, acc.: %.2f%%] [G loss: %f]" % (epoch, d_loss[0], 100*d_loss[1], g_loss))
              print("Epoch: "+str(epoch)+" | Discriminator Loss: "+ str(d_loss[0])+ " | Discriminator Accuracy: "+ str(100*d_loss[1])+" | Generator Loss: "+str(g_loss[0])+ " | Generator Accuracy: "+str(100*g_loss[1]))
              D_losses.append(d_loss[0])
              G_losses.append(g_loss[0])
              D_acc.append(100*d_loss[1])
              G_acc.append(100*g_loss[1])
```
